```python
import os 
from netCDF4 import Dataset as NetCDFFile
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature
import pickle
import matplotlib as mpl
```

```python
nc = NetCDFFile("ensmean_CMIP6_historical_minus_histnat_1985_2014.nc")
lats = nc.variables['lat'][:]
lons = nc.variables['lon'][:]
pr   = nc.variables['pr'][:]
nc.close()
nc = NetCDFFile("enspctl75_CMIP6_historical_minus_histnat_1985_2014.nc")
pr_75   = nc.variables['pr'][:]
nc.close()
nc = NetCDFFile("enspctl25_CMIP6_historical_minus_histnat_1985_2014.nc")
pr_25   = nc.variables['pr'][:]
nc.close()

```

```python
from mpl_toolkits.basemap import Basemap
# Get some parameters for the Stereographic Projection
lon_0 = lons.mean()
lat_0 = lats.mean()

m = Basemap(width=6000000,height=4500000,
            resolution='l',projection='stere',\
            lat_ts=40,lat_0=lat_0,lon_0=lon_0)
```

```python
# Because our lon and lat variables are 1D,
# use meshgrid to create 2D arrays
# Not necessary if coordinates are already in 2D arrays.
lon, lat = np.meshgrid(lons, lats)
xi, yi = m(lon, lat)
```

```python
v =  np.linspace(-0.5,0.5 , 41, endpoint=True)
    #cmap = plt.cm.  # define the colormap
cmap = plt.cm.seismic
    # extract all colors from the .jet map
cmaplist = [cmap(i) for i in range(cmap.N)]
#    # force the first color entry to be grey
#    cmaplist[0] = (.5, .5, .5, 1.0)

    # create the new map
cmap = mpl.colors.LinearSegmentedColormap.from_list('Custom cmap', cmaplist, cmap.N)
norm = mpl.colors.BoundaryNorm(v, cmap.N)
```

```python
# Plot Data
fig = plt.figure('1')
fig.set_size_inches(14, 10)
cs = m.pcolor(xi,yi,np.squeeze(pr), vmin=-.5, vmax=.5,cmap=cmap, norm=norm)
# Add Grid Lines
m.drawparallels(np.arange(-80., 81., 10.), labels=[1,0,0,0], fontsize=10)
m.drawmeridians(np.arange(-180., 181., 10.), labels=[0,0,0,1], fontsize=10)

# Add Coastlines, States, and Country Boundaries
m.drawcoastlines()
m.drawstates()
m.drawcountries()

# Add Colorbar
cbar = m.colorbar(cs, location='bottom', pad="10%")
cbar.set_label("[mm/day]")

# Add Title
plt.title("CMIP6 ensembles' mean precipitation change  \n historical minus histnat (162 simulations) \n 1985-2014")
plt.savefig("CMIP6_ensmean_historical_minus_histnat.png", bbox_inches='tight')
plt.show()
```

```python
# Plot Data
fig = plt.figure('1')
fig.set_size_inches(14, 10)
cs = m.pcolor(xi,yi,np.squeeze(pr_25), vmin=-.5, vmax=.5,cmap=cmap, norm=norm)

#m.plot(xi[(pr<pr_75) ],yi[(pr<pr_75) ],'k.')
# Add Grid Lines
m.drawparallels(np.arange(-80., 81., 10.), labels=[1,0,0,0], fontsize=10)
m.drawmeridians(np.arange(-180., 181., 10.), labels=[0,0,0,1], fontsize=10)

# Add Coastlines, States, and Country Boundaries
m.drawcoastlines()
m.drawstates()
m.drawcountries()

# Add Colorbar
cbar = m.colorbar(cs, location='bottom', pad="10%")
cbar.set_label("[mm/day]")

# Add Title
plt.title("CMIP6 ensembles' $25^{th}$ percentile precipitation change  \n historical minus histnat (162 simulations) \n 1985-2014")
plt.savefig("CMIP6_enspctl25_historical_minus_histnat.png", bbox_inches='tight')

plt.show()
```
